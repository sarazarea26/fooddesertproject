```python
import pandas as pd
```

```python
# filter by state and county (Dallas County, TX)

df = pd.read_csv("snap_retailer_location_data.csv")
dallas_snap = df[(df["State"]== "TX") & (df["County"]=="DALLAS")]

print(dallas_snap.shape)
print(dallas_snap["Store_Type"].value_counts())
```

```python
# what is in the other cateogry
dallas_snap[dallas_snap["Store_Type"] == "Other"]["Store_Name"].unique()
```

```python
# want to exclude Dollar General, Dollar Tree, CVS Pharmacy, CVS, Family Dollar,  Walgreens

excluded_others=["Dollar", "CVS", "Walgreens", "Pharmacy"]

new_data = dallas_snap[(dallas_snap["Store_Type"] == "Other") 
    & ~dallas_snap["Store_Name"].str.contains("|".join(excluded_others), case=False, na=False)].copy()
print(new_data.shape)
print(new_data["Store_Name"].value_counts())

# grocery stores i excluded from the list: "Wrehsonco", "Abyssinia Market No", "J DAVIS FISH MARKET", "TEXAS SEAFOOD FISH MARKET", "Chuchu Grocery & Catering", "AZ Food Grocery", "Star Medical Specialties", "Wohemiya Ethiopian Grocery", "1 Star Beverage", "Labaf African Market", "Dukem Ethiopian Grocery And Restaurant", "El Mayoreo Meat Market", "Star Food Mart", "Hiwot Injera", "Flavor Spice", "Zion African Market NO", "Judy Ethiopian Grocery", "Al Markaz", "Deshi Bazzar Inc", "New Season 24", "Ac Injera & Grocery", "     
```

```python
# filter out stores that are not grocery stores
included = ["Super Store","Supermarket","Grocery Store", "Farmers and Markets"]
keep_from_others = [
    "Kelley Produce Dfm", "Richland Market Grocery Inc.", "La Michoacana Meat Market 28",
    "Wrights Family Food", "LA AZTECA MEAT MARKET 6", "Afribbean Market 2",
    "INDIAN SUPERSTORE 1", "Pleasant Grove Meat Market", "Trippleg African Food Mart Llc",
    "Afghan Asian Halal Market", "Don Cuco Meat Market Don Cuco#3", "La Michoacana Supermarket #28",
    "Farmer's Delite 001", "Uac International Food", "Deshi Bazzar Inc",
    "India Bazaar", "Mango Tropical", "Jf African & Asian Market"
]
cleaned_snap_data = dallas_snap[dallas_snap["Store_Type"].isin(included) | 
    dallas_snap["Store_Name"].isin(keep_from_others)| 
    dallas_snap["Store_Name"].str.contains("Braum", case=False, na=False)].copy()

# to check and make sure it is working
cleaned_snap_data[cleaned_snap_data["Store_Name"].str.contains("Braum", case=False, na=False)]
```

```python
# remove columns that are not needed :"Additonal_Address", "Incentive_Program", "Grantee_Name", "X", "Y"

final_cleaned_snap_data = cleaned_snap_data.drop(columns=["Additonal_Address", "Incentive_Program", "Grantee_Name", "X", "Y", "Zip4"])
```

```python
# Check for duplicate store names at the same address
dupes = final_cleaned_snap_data[cleaned_snap_data.duplicated(subset=["Store_Name", "Store_Street_Address"], keep=False)]
print(dupes[["Store_Name", "Store_Street_Address"]])
```

```python
# final csv file created
final_cleaned_snap_data.to_csv("cleaned_snap_data.csv", index=False)
print(final_cleaned_snap_data.shape)
print(final_cleaned_snap_data[["Store_Name", "Latitude", "Longitude", "Store_Type"]].head())
```
