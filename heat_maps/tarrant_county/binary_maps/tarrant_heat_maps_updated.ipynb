# 1. Median Income Map
Imported Packages

```python
import pandas as pd
```

```python
df = pd.read_csv('tarrant_income.csv')
```

```python
import shapely
import numpy as np
import fiona
import geopandas as gpd
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = "retina"

import os
os.listdir()
```

SHP File

```python
gdf = gpd.read_file("tl_2024_48_tract.shp")
```

```python
gdf.head()
```

Tarrant County

```python
tarrant = gdf[gdf["COUNTYFP"] == "439"]
tarrant
```

Merging Median Income and Map Files

```python
merged = tarrant.merge(
    df,
    left_on = "GEOIDFQ",
    right_on = "GEO_ID",
    how = "left"
)
merged
```

```python
merged["S1903_C03_001E"] = pd.to_numeric(
    merged["S1903_C03_001E"],
    errors="coerce"
)
```

```python
fig, ax = plt.subplots(figsize=(10,10))

merged.plot(
    column = "S1903_C03_001E",
    cmap = "viridis",
    legend = True,
    ax=ax
) 

plt.title("Median Household Income by Census Tract in Tarrant County")
plt.axis("off")
plt.show()
```
