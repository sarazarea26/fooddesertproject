```python
import pandas as pd
```

```python
# filter by state and county 

df = pd.read_csv("snap_retailer_location_data.csv")
collin_snap = df[(df["State"]== "TX") & (df["County"]=="COLLIN")]

print(collin_snap.shape)
print(collin_snap["Store_Type"].value_counts())
```

```python
# what is in the other cateogry
collin_snap[collin_snap["Store_Type"] == "Other"]["Store_Name"].unique()
```

```python
# want to exclude Dollar General, Dollar Tree, CVS Pharmacy, CVS, Family Dollar,  Walgreens

excluded_others=["Dollar", "CVS", "Walgreen", "Walgrens", "Pharmacy"]

new_collin_data = collin_snap[(collin_snap["Store_Type"] == "Other") 
    & ~collin_snap["Store_Name"].str.contains("|".join(excluded_others), case=False, na=False)].copy()
print(new_collin_data.shape)
print(new_collin_data["Store_Name"].unique())
```
