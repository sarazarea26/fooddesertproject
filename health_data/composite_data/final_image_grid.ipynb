```python
from PIL import Image

images = [
    Image.open("Dallas_Composite_Map.png"),
    Image.open("Harris_Composite_Map.png"),
    Image.open("Travis_Composite_Map.png"),
    Image.open("Tarrant_Composite_Map.png"),
    Image.open("Bexar_Composite_Map.png")
]

# resizing the images so that they are still legible 
thumb_w, thumb_h = 450, 325
resized_images = [img.resize((thumb_w, thumb_h), Image.Resampling.LANCZOS) for img in images]

# creating a new image background 
grid_canvas = Image.new("RGB", (thumb_w * 3, thumb_h * 2), "white")

positions = [
    (0, 0),             # Row 1, Col 1 (Dallas)
    (thumb_w, 0),       # Row 1, Col 2 (Harris)
    (thumb_w * 2, 0),   # Row 1, Col 3 (Travis)
    (0, thumb_h),       # Row 2, Col 1 (Tarrant)
    (thumb_w, thumb_h), # Row 2, Col 2 (Bexar)
]

for img, pos in zip(resized_images, positions):
    grid_canvas.paste(img, pos)

grid_canvas.show()
grid_canvas.save("all_metros_health_grid.png", dpi=(300, 300))
```
