```python
import pandas as pd
df = pd.read_csv("PLACES__Census_Tract_Data_(GIS_Friendly_Format),_2025_release_20260624.csv")
```

```python
county = "Bexar"
```

```python
# filter by location 
filtered_data = df[
    (df["StateAbbr"] == "TX") & 
    (df["CountyName"] == f"{county}")
][["StateAbbr", "CountyName", "TractFIPS", "CHD_CrudePrev", "CHD_Crude95CI"]]
filtered_data
```

```python
# create a new csv
filtered_data.to_csv(f"{county}_CHD_data.csv", index=False)
```
