```python
import geopandas as gpd
```

```python
gdf = gpd.read_file("../usda_ofm.gpkg")
gdf
```

```python
gdf.columns
```

```python
for col in gdf.columns:
    print(col)
```

```python
gdf[["location_x", "location_y"]].head()
```

```python
df = gdf[[
    "listing_name",
    "location_address",
    "location_y",
    "location_x",
    "OID",
    "GEOID",
    "STATE",
    "COUNTY"
]].copy()

df = df.rename(columns={
    "listing_name": "name",
    "location_address": "address",
    "location_y": "latitude",
    "location_x": "longitude",
    "STATE": "state",
    "COUNTY": "county",
    "OID": "oid",
    "GEOID": "geoid"
})
```

```python
df.to_csv("usda_ofm.csv")
```

```python
gdf1 = gpd.read_file("../dallasFarmsFinal.geojson")
```

```python
print(gdf1.columns)
```

```python
gdf1["longitude"] = gdf1.geometry.x
gdf1["latitude"] = gdf1.geometry.y

df = gdf1[[
    "SiteName",
    "Address",
    "latitude",
    "longitude"
]].copy()

df = df.rename(columns={
    "SiteName": "name",
    "Address": "address"
})

df["state"] = None
df["county"] = None

df.to_csv("cleaned_dallas_geojson.csv", index=False)
```
