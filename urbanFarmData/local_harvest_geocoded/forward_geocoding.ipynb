```python
import googlemaps
import os
from dotenv import load_dotenv
import pandas as pd
import time 

load_dotenv()

api_key = os.getenv("google_api_key")

def geo_code_address(address: str) -> dict:
    result = gmaps.geocode(address)
    if result:
        location = result[0]["geometry"]["location"]
        return {
            "lat": location["lat"],
            "lon": location["lng"]
        }
    else:
        return {"lat": None, "lon": None}

city = "san_antonio"
df = pd.read_csv(f"../{city}_localharvest.csv")

for index, row in df.iterrows():
    if pd.notna(row["street_address"]) and row["street_address"] != "Not found" \
       and (pd.isna(row["latitude"]) or pd.isna(row["longitude"])):
        
        address = row["street_address"].replace("Location: ", "").strip()
        print(f"Geocoding: {address}")
        coords = geo_code_address(address)
        print(f"  Result: {coords}")
        df.at[index, "latitude"] = coords["lat"]
        df.at[index, "longitude"] = coords["lon"]
        
        time.sleep(0.1)

df.to_csv(f"{city}_geocoded.csv", index=False)
```
