```python
import pandas as pd
```

```python
county = "bexar"

df = pd.read_csv(f"../cleaned_boundary_data/cleaned_boundary_{county}_farms.csv")

pattern = (
    r"farmers'?\s*market|nursery|market\s*place|flea\s*market|florist|bees?|"
    r"mart|ranch|market\s*hall|market\s*center|trade\s*center|supply|grocers|"
    r"rose\s*garden|orchid\s*house|designs\s*&\s*events|dallas\s*county|warehouse|playcenter"
)
df = df[~df["name"].str.contains(pattern, case=False, na=False, regex=True)]

# Save the filtered result to a new CSV
df.to_csv(f"filtered_{county}_farms.csv", index=False)
```
